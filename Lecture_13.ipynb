```python
import pandas as pd

df = pd.read_csv('./sonar3.csv', header=None)

df.head(5)
```

```python
# 마지막 열이 광물의 종류를 나타냄
df[60].value_counts()
```

```python
x = df.iloc[:, 0:60]
y = df.iloc[:,60]
```

```python
from tensorflow import keras
from keras import Sequential, Input
from keras.layers import Dense

model = Sequential()
model.add(Input(shape=(60,)))
model.add(Dense(24, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
history = model.fit(x, y, epochs=200, batch_size=10, verbose=2)
```

### **과적합** 발생

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True)
```

```python
model = Sequential()
model.add(Input(shape=(60,)))
model.add(Dense(24, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
history = model.fit(x_train, y_train, epochs=200, batch_size=10, verbose=2)
```

```python
score = model.evaluate(x_test, y_test)
print('Test Accuracy %.2f%%' %(score[1]*100))
print(model.evaluate(x_test, y_test))
```

### k겹 교차 검증 (k-Fold Cross-Validation)

```python
from sklearn.model_selection import KFold

k = 5
kfold = KFold(n_splits=k, shuffle=True)
acc_score=[]

for train_index, test_index in kfold.split(x):
  x_train, x_test = x.iloc[train_index,:], x.iloc[test_index,:]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
model = Sequential()
model.add(Input(shape=(60,)))
model.add(Dense(24, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs=200, batch_size=10, verbose=2)
```

```python
accuracy = model.evaluate(x_test, y_test, verbose=2)[1]
acc_score.append(accuracy)
```

```python
avg_acc_score = sum(acc_score) / k
```

```python
print('정확도:', acc_score)
print('평균 정확도: %.2f%%' % (avg_acc_score * 100))
```
