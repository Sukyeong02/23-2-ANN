### 파이썬으로 선형회귀 귀현하기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

plt.figure()
plt.scatter(x,y,marker='o',c='k',s=100)
plt.xlabel('Study Time(x)')
plt.ylabel('Score(y)')a
plt.grid(True)
plt.show()
```

```python
a = 0
b = 0

lr = 0.03

epoch = 2001

n = len(x)

for j in range(epoch):
    y_pred = a * x + b
    error = y - y_pred

    a_diff = (2/n) * sum(-x * error)
    b_diff = (2/n) * sum(-error)

    a = a - lr * a_diff
    b = b - lr * b_diff

    if (j % 100) == 0:
        print('epoch=%.f, 기울이=%.4f, y절편=%.4f' % (j,a,b))

y_pred = a * x + b

plt.figure()
plt.scatter(x,y,marker='o',c='k',s=100)
plt.plot(x,y_pred,marker='*',c='r',markersize=15)
plt.xlabel('Study Time(x)')
plt.ylabel('Score(y)')
plt.grid(True)
plt.show()
```

### 다중선형회귀

```python
x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3]) #과외시간
y = np.array([81,93,91,97])
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection="3d")
ax.scatter3D(x1, x2, y, s=200)
ax.set_xlabel("Study Time(x1)")
ax.set_ylabel("Private Lesson Time(x2)")
ax.set_zlabel("Score(y)")
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.01 #0.03은 발산함

epoch = 2001

n = len(x1)

for j in range(epoch):
    y_pred = a1 * x1 + a2 * x2 + b
    error = y - y_pred

    a1_diff = (2/n) * sum(-x1 * error)
    a2_diff = (2/n) * sum(-x2 * error)
    b_diff = (2/n) * sum(-error)

    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff

    if (j%100) == 0:
        print('epoch=%.f, 기울기1=%.4f, 기울기2=%.4f, y절편=%.4f' %(j,a1,a2,b))

print('실제 점수:', y)
print('예측 점수:', y_pred)
```
