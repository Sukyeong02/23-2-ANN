```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential, Input
from keras.layers import Dense
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
```

```python
model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(1, activation=('linear')))

model.compile(optimizer='sgd', loss='mse')   # sgd = 경사하강법, mse = 평균제곱오차

model.fit(x,y,epochs=2000)
```

```python
y_pred = model.predict(x)

plt.figure()
plt.scatter(x, y, marker='o', c='k', s=100)
plt.plot(x, y_pred, marker='*', c='r', markersize=15)
plt.xlabel('Study Time(x)')
plt.ylabel('Score(y)')
plt.grid(True)
plt.show()

hour = 10
prediction = model.predict([hour])

print('%.f시간을 공부할 경우의 예상 점수는 %.2f점입니다.' %(hour, prediction))
```

```python
x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])
```

```python
model = Sequential()
model.add(Input(shape=(2,)))
model.add(Dense(1, activation='linear'))

model.compile(loss='mse', optimizer='sgd')

model.fit(x, y, epochs = 2000)

hour = 7
private_class= 3
result = model.predict([[hour, private_class]])

print('%.f시간을 공부하고 %.f시간의 과외를 받는 경우, 예상 점수는 %.f입니다.' %(hour, private_class, result))
```
